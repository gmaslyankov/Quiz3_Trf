# Merton

# Merton with jumps

```python
from credit_spreads import credit_spread_Merton_jump
```

```python
time = np.linspace(0.01, 10, 100)
for lambd in [0.01, 0.05, 0.2]:
    plt.plot(time, [credit_spread_Merton_jump(100, 70, t, 0.3, 0.05, lambd, -0.2, 0.2)[0] for t in time], label=f'lambda={lambd}')

plt.legend()
```

```python
time = np.linspace(0.01, 10, 100)
for mu_J in [-0.1, -0.3, -0.5]:
    plt.plot(time, [credit_spread_Merton_jump(100, 70, t, 0.3, 0.05, 0.05, mu_J, 0.2)[0] for t in time], label=f'mu J={mu_J}')

plt.legend()
```
